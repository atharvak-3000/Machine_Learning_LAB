#Linear Regression
-- predicting backache risk score

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#sample data
X = np.array([2, 3, 4, 5, 1]).reshape(-1,1)
Y = np.array([1.2, 2.3, 4.0, 5.0, 2.0])
```

```python
#scatterplot
plt.scatter(X,Y)
plt.title("Driving Hours VS Backache Risk Score")
plt.xlabel("Hours")
plt.ylabel("Backache")
```

```python
#Training Model
model = LinearRegression()
model.fit(X,Y)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Y = {slope:.2f} * X + {intercept:.2f}")
```

```python
y_pred = model.predict(X)
```

```python
#scatterplot
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')

# Titles and labels
plt.title("Driving Hours VS Backache Risk Score")
plt.xlabel("Hours")
plt.ylabel("Backache Risk Score")
plt.legend()
plt.grid(True)
plt.show()
```
